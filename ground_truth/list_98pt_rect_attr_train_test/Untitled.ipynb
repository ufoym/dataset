```python
with open('list_98pt_rect_attr_test.txt', 'r') as f:
    lines = f.readlines()
```

```python
info = {}
for line in lines:
    terms = line.split()
    xys = [float(t) for t in terms[:-11]]
    rect = [int(t) for t in terms[-11:-7]]
    attrs = [int(t) for t in terms[-7:-1]]
    fn = terms[-1].strip()
    info[fn] = (xys, rect, attrs)
```

```python
import os
import numpy as np
paths, gtruths = [], []
with open('/root/liteface/var/dataset/wflw/ground_truth/list.txt', 'r') as f:
    for line in f.readlines():
        terms = line.strip().split()
        path = terms[0]
        lmks_gt = np.asarray(terms[1:197], dtype=np.float)
        pose_gt = np.asarray(terms[203:206], dtype=np.float)
        type_flag = np.asarray(terms[206], dtype=np.int)
        if type_flag == -2:
            paths.append(path)
            gtruths.append((lmks_gt, pose_gt, type_flag))
```
